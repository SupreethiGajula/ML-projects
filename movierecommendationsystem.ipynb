```python
df1.columns = ['user_id','item_id','rating','timestamp']
df1.head()
```

```python
df2 = pd.read_csv("C:/Users/gajul/OneDrive/Desktop/MLprojects/datasets/Movie_Id_Titles")
```

```python
df2.head()
```

```python
#merging movie ratings and movie titles
df = pd.merge(df1,df2, on = 'item_id')
df.head()
```

```python
#grouping same movie entries
#we are sorting by taking the mean of ratings given by different users
ratings_and_no_of_ratings = pd.DataFrame(df.groupby('title')['rating'].mean().sort_values(ascending = False))
ratings_and_no_of_ratings
```

```python
ratings_and_no_of_ratings['no_of_ratings'] = df.groupby('title')['rating'].count()
ratings_and_no_of_ratings
```

```python
ratings_and_no_of_ratings = ratings_and_no_of_ratings.sort_values('no_of_ratings',ascending = False)
```

```python
ratings_and_no_of_ratings
```

```python
#creating a pivot table with ratings as values
pt = df.pivot_table(index='user_id',columns='title',values='rating')
pt.head()
```

```python
#now printing all the movie names we have
ratings_and_no_of_ratings.index
```

```python
#now testing with a live prediction
#taking the movie name as input
#finding the test movie name in the pivottable pt
#corrwith() function correlates or finds the similar movies to our test movie
#then we create a dataframe corr_df with a single column correlation
#then we join the corr_df with the no_of_ratings column from the ratings_and_no_of_ratings dataframe
#now we sort the values based on the correlation 

test = input("\nenter the movie name\n")
movie_vector  =  pt[test].dropna()
similar_movies = pt.corrwith(movie_vector)
corr_df = pd.DataFrame(similar_movies,columns=['Correlation'])
corr_df = corr_df.join(ratings_and_no_of_ratings['no_of_ratings'])
corr_df = corr_df[corr_df['no_of_ratings']>100].sort_values('Correlation',ascending=False).dropna()
corr_df.head(10)
```
